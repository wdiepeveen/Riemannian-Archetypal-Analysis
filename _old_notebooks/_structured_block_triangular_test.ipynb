```python
import torch

from src.transforms.vector.linear.structured_block_upper_triangular import StructuredBlockUpperTriangularVectorTransform
from src.transforms.vector.linear.structured_block_lower_triangular import StructuredBlockLowerTriangularVectorTransform
```

```python
transform = StructuredBlockUpperTriangularVectorTransform(d=100, r=5, bias=True, random_init=True, efficient_inverse=True)
input_matrix = torch.randn(10, 100)

print("weight_11:", transform.weight_11)
print("weight_12:", transform.weight_12)
print("bias:", transform.bias)

output_matrix, log_abs_det = transform(input_matrix)
print("Output matrix shape:", output_matrix.shape)  # should be [10, 6]
print("Output matrix:", output_matrix)
print("Log absolute determinant:", log_abs_det)

# test inverse
inverse_transform, log_abs_det_inv = transform.inverse(output_matrix)
print("Inverse transform shape:", inverse_transform.shape)  # should be [10, 6]
print("Inverse transform:", inverse_transform)
print("Log absolute determinant (inverse):", log_abs_det_inv)

# test reconstruction
print("Reconstruction error:", torch.norm(input_matrix - inverse_transform).item())
```
