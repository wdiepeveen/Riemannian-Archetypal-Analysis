```python
import torch
import torch.nn as nn

from src.transforms.image.parity_conv import ParityConvImageTransform
from src.transforms.image.parity_conv.linear_2d import Linear2DParityConvImageTransform
from src.transforms.image.parity_conv.simple_tanh_2d import SimpleTanh2DParityConvImageTransform
```

```python
x = torch.randn(16, 3, 32, 32)
```

```python
transform = ParityConvImageTransform(in_channels=3, height=32, width=32, conv_network=nn.Conv2d(3, 3, kernel_size=3, padding=1), parity=1)
z, log_det = transform.forward(x)
x_recon, log_det_inv = transform.inverse(z)

assert torch.allclose(x, x_recon, atol=1e-6), "Inverse transformation did not reconstruct the original input accurately."
assert torch.allclose(log_det, -log_det_inv, atol=1e-6), "Log determinants do not match for forward and inverse transformations."
```
