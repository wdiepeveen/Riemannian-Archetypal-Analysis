```python
import torch

from src.diffeomorphisms.vector.product_embedding.poincare_identity import PoincareIdentityVectorDiffeomorphism
from src.diffeomorphisms.vector.structured_linear import StructuredLinearVectorDiffeomorphism
```

```python
# test Poincare-Identity diffeomorphism
x = torch.randn(10, 3)
phi = PoincareIdentityVectorDiffeomorphism(3, 2)
phi_x = phi.forward(x)
x_reconstructed = phi.inverse(phi_x)
print("Max reconstruction error of Poincare-Identity diffeomorphism:", (x - x_reconstructed))
```

```python
# test structure linear diffeomorphism
W = torch.randn(2, 3)
phi_linear = StructuredLinearVectorDiffeomorphism(W)
phi_x_linear = phi_linear.forward(x)
x_reconstructed_linear = phi_linear.inverse(phi_x_linear)
print("Max reconstruction error of Structured Linear diffeomorphism:", (x - x_reconstructed_linear))
```
